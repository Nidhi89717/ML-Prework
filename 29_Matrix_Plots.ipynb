```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('gdrive/My Drive/csv_files/country_table.csv')
```

```python
df = df.set_index('Countries')
```

## Heatmap

```python
sns.heatmap(df)
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1))
```

```python
sns.heatmap(df.drop('Life expectancy',axis=1),lw=0.5)
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy',axis=1),lw=0.5,annot=True)
```

```python
#Note its not palette here
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy',axis=1),lw=0.5,annot=True,cmap='viridis')
```

```python
# Set colorbar based on value from dataset
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy',axis=1),lw=0.5,annot=True,cmap='viridis',center=40)
```

## Clustermap

Plot a matrix dataset as a hierarchically-clustered heatmap.

```python
plt.figure(dpi=200)
sns.clustermap(df.drop('Life expectancy',axis=1),lw=0.5,annot=True,cmap='viridis')
```

```python
plt.figure(dpi=200)
sns.clustermap(df.drop('Life expectancy',axis=1),lw=0.5,annot=True,cmap='viridis',col_cluster=False)
```
