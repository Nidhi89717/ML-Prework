## catplot()

Kind Options are: “point”, “bar”, “strip”, “swarm”, “box”, “violin”, or “boxen”

```python
sns.catplot(data=df,x='gender',y='math score',kind='box')
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',row='lunch')
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',col='lunch')
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',row='test preparation course',col='lunch');
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',row='race/ethnicity',col='lunch');
```

```python
sns.catplot(data=df,x='gender',y='math score',kind='violin',row='test preparation course',col='lunch');
```

## Pairgrid

```python
sns.PairGrid(df)
```

```python
g = sns.PairGrid(df)

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.histplot)

g = g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(df,hue='gender',palette='Set2')

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.kdeplot)

g = g.map_lower(sns.kdeplot)

g = g.add_legend()
```

```python
g = sns.PairGrid(df, hue="gender", palette="viridis",hue_kws={"marker": ["o", "+"]})
g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.distplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend();

# Safely ignore the warning, its telling us it didn't use the marker for kde plot
```
