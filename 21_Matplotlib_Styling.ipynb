# Matplotlib Styling

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,11)
```

## Legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')
ax.plot(x, x**2,label='X vs X^2')
ax.legend(loc=0);
```

ax.legend(loc=1) - upper right corner <br>
ax.legend(loc=2) - upper left corner <br>
ax.legend(loc=3) - lower left corner <br>
ax.legend(loc=4) - lower right corner <br>

 .. many more options are available

**Most common to choose**<br>
ax.legend(loc=0) - let matplotlib decide the optimal location

```python
ax.legend(loc=(1.5,0.5));
fig
```

### Colors with the color parameter

We can also define colors by their names or RGB hex codes and optionally provide an alpha value using the `color` and `alpha` keyword arguments. Alpha indicates opacity.

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x,color='#e60ee2',label='X Vs X')
axes.plot(x,x+1,color='blue',label='X vs X+1')
axes.legend();
```

With matplotlib, we can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where `'b'` means blue, `'g'` means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, 'b.-' means a blue line with dots:

```python
fig,axes = plt.subplots(figsize=(8,6))

axes.plot(x,x**2,'b--')
axes.plot(x,x**3,'g.-')
```

## Line and marker styles

### Linewidth

To change the line width, we can use the `linewidth` or `lw` keyword argument. 



### Linestyles

There are many linestyles to choose from, here is the selection:<br>

** possible linestype options --, , -., :, steps**

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x,color='#e6c017',lw=5,ls='--')
axes.plot(x,x+1,color='purple',linewidth=2,linestyle='-.');
```

### Custom linestyle dash

The dash sequence is a sequence of floats of even length describing
the length of dashes and spaces in points.

For example, (5, 2, 1, 2) describes a sequence of 5 point and 1 point
dashes separated by 2 point spaces.

First, we see we can actually "grab" the line from the .plot() command

```python
fig,axes = plt.subplots(figsize=(8,6))

lines = axes.plot(x,x+1)

type(lines)
```

```python
fig,axes = plt.subplots(figsize=(8,6))

line = axes.plot(x,x+8,color='black',lw=5)
line[0].set_dashes([10,2]) # format: line length, space length
```

```python
fig,axes = plt.subplots(figsize=(8,6))
lines = axes.plot(x,x+8,color='black',lw=5)
lines[0].set_dashes([1,1,2,2,10,5])
```

## Markers



### Markers Style

Huge list of marker types can be found here: https://matplotlib.org/3.2.2/api/markers_api.html

```python
fig,ax = plt.subplots(figsize=(8,6))

ax.plot(x,x-1,marker='+', markersize=20)
ax.plot(x,x+2,marker='o',ms=20)
ax.plot(x,x-3,marker='s',ms=20,lw=0)
ax.plot(x,x+4,marker='1',ms=20);
```

### Custom marker edges, thickness,size,and style

```python
fig,axes = plt.subplots(figsize=(8,6))

axes.plot(x,x-2,marker='s',markersize=20,markerfacecolor='red',markeredgewidth=8,markeredgecolor='blue');
```
