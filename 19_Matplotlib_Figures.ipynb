# Matplotlib Figure Object

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creating a Figure 

**NOTE: ALL THE COMMANDS NEED TO GO IN THE SAME CELL!**

To begin create a figure instance. Then add axes to that figure

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(a,b);
```

### Adding another set of axes to the Figure

```python
fig = plt.figure()

#Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

#Small Axes
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1]) # Full figure
axes2 = fig.add_axes([0.2,0.5,0.25,0.25]) # Small figure
axes3 = fig.add_axes([1,1,0.25,0.25]) # Starts at top right corner!

axes1.plot(a,b)
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('Big figure')

axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,8000)
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes2.set_title('Zoomed In')

axes3.plot(a,b);
```
