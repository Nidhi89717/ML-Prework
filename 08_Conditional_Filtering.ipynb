## Conditions

```python
df['total_bill'] > 40
```

```python
bool_ser = df['total_bill'] > 40
```

```python
df[bool_ser]       # Return the rows where the boolean expression is True
```

```python
df[df['sex'] == 'Male']
```

## Multiple Conditions

AND ---> & <br>
OR  ---> | <br>

```python
df[(df['total_bill'] > 30) | (df['sex'] == 'Male')]
```

```python
df[(df['day'] == 'Sun') | (df['day'] == 'Sat')]
```

##Conditional Operators isin()

Use.isin() operators to filter by a list of options

```python
options = ['Sun', 'Sat']
df['day'].isin(options)
```

```python
df[df['day'].isin(['Sun', 'Sat'])]
```
