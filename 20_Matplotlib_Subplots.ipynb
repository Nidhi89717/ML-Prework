# Matplotlib Sub Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

# plt.subplots()

```python
fig, axes = plt.subplots(nrows=1, ncols = 2)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
  ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)
axes[1][1].plot(a,b)

plt.tight_layout() #automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:
```

## Parameters on subplots()

```python
fig,axes = plt.subplots(nrows=2, ncols=2, figsize=(8,8))

axes[0][0].plot(a,b)
axes[0][0].set_title('Title 0,0')

axes[0][1].plot(x,y)
axes[0][1].set_xlabel('Xlabel 0,1')
axes[0][1].set_title('0,1 Title')

axes[1][0].plot(b,a)

axes[1][1].plot(y,x)

fig.suptitle('Figure Level', fontsize = 16)
```

### Manual spacing on subplots()

* left = 0.125  # the left side of the subplots of the figure
* right = 0.9   # the right side of the subplots of the figure
* bottom = 0.1  # the bottom of the subplots of the figure
* top = 0.9     # the top of the subplots of the figure
* wspace = 0.2  # the amount of width reserved for space between subplots,
              # expressed as a fraction of the average axis width
* hspace = 0.2  # the amount of height reserved for space between subplots,
              # expressed as a fraction of the average axis height

```python
fig,axes = plt.subplots(nrows=2, ncols=2, figsize = (12,8))

axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(y,x)
axes[1][1].plot(b,a)

fig.subplots_adjust(wspace=0.9, hspace=1)
```
