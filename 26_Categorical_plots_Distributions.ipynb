## Boxplot

A boxplot display distribution through the use of quartiles and an IQR for outliers.

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,y='math score');
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df,y='math score',x='test preparation course');
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.boxplot(data=df, y='math score',x='parental level of education',hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,0.5));
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.boxplot(data=df, x='math score',y='parental level of education',hue='test preparation course',palette='Set2')
plt.legend(bbox_to_anchor = (1.05,0.5));
```

## Violinplot

A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df, x='math score',y='parental level of education');
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df, x='math score',y='parental level of education',hue='test preparation course',palette='Set2')
plt.legend(bbox_to_anchor = (1.35,0.5));
```

### Violinplot Parameters

#### split
When using hue nesting with a variable that takes two levels, setting split to True will draw half of a violin for each level. This can make it easier to directly compare the distributions.

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df, x='math score',y='parental level of education',hue='test preparation course',palette='Set1',split=True)
plt.legend(bbox_to_anchor = (1.25,0.5));
```

#### inner

Representation of the datapoints in the violin interior. If box, draw a miniature boxplot. If quartiles, draw the quartiles of the distribution. If point or stick, show each underlying datapoint. Using None will draw unadorned violins.

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df, x='math score',y='parental level of education',hue='test preparation course',palette='Set1',split=True,inner=None)
plt.legend(bbox_to_anchor = (1.35,0.5));
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df, x='math score',y='parental level of education',inner='quartile');
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df, x='math score',y='parental level of education',inner='stick');
```

#### bandwidth

Similar to bandwidth argument for kdeplot

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df, x='math score',y='parental level of education',bw=0.1);
```

```python
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(data=df, x='math score',y='parental level of education',bw=0.01);#Noisy
```
